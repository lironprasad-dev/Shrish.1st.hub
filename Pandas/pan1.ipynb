```python
dic2={
    "Students":["student1",
    "student2",
    "student2",
    "student3",
    "student5"],
    "Roll":['1','6','6',np.nan,'7'],
    "Gender":["Feale","Female","Female",np.nan,"Male"],
    "Marks Obtained":[22,67,67,np.nan,71]
}
dic2
```

```python
dic2_df=pd.DataFrame(dic2)
dic2_df
```

```python
dic2_df.isnull().sum()
```

```python
dic2_df.duplicated().sum()
```

### deleting duplicated values


```python
dic2_df.drop_duplicates()
```

### dropping null/missing values


```python
dic2_df.dropna()
```

### mean and mode imputation

```python
dic2_df
```

```python
gender_mode=dic2_df['Gender'].mode()[0]
gender_mode
```

```python
marks_mean=dic2_df['Marks Obtained'].mean()
marks_mean
```

### filling missing values

```python
dic2_df['Gender']=dic2_df['Gender'].fillna(gender_mode)
dic2_df['Marks Obtained']=dic2_df['Marks Obtained'].fillna(marks_mean)
dic2_df
```

```python
dic2_df=dic2_df.drop_duplicates()
dic2_df=dic2_df.dropna()
dic2_df
```
