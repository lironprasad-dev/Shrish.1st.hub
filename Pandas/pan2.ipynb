### checking missing values

```python
df.isnull().sum()
```

```python
df["Embarked"].value_counts()
```

```python
age_mean=df["Age"].mean().__round__()
embarked_mode=df["Embarked"].mode()[0]
```

```python
df["Age"]=df["Age"].fillna(age_mean)
df["Embarked"]=df["Embarked"].fillna(embarked_mode)
df.head(10)
```

```python
df.isnull().sum()
```

```python
df["Survived"].value_counts()
```

### removing unimportant features(columns)

```python
removing_columns= ["PassengerId","Name","Ticket"]
df=df.drop(columns=removing_columns)
df.head(10)
```

```python
df["Family_Size"]=df["SibSp"] + df["Parch"] +1
removing_columns= ["SibSp","Parch"]
df=df.drop(columns=removing_columns)
df=df.drop(columns=["Cabin"])
df.head(4)
```

### converting categorical data into numerical

```python
df["Sex"]=df["Sex"].map({"male":1,"female":0})
df.head(5)
```

```python
df1=df.copy()
df1=pd.get_dummies(df1, columns=['Embarked'],dtype=int, drop_first=True)
df1.head(10)
```

```python
def fair_category(fare):
    if fare<15 :
        return ("Low")
    elif fare>40:
        return("Medium")
    else:
        return("High")
    
df1["Fair_Category"]=df["Fare"].apply(fair_category)
df1.head(10)
```
